## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision this improves: Which pages should an editor review first to improve CTR or engagement — rewriting titles, fixing meta descriptions, improving snippets, or adding content.

Who acts: A content editor or SEO manager with limited time. They open the ranked list and pick the top 20–50 pages to work on this week.

Cost of a wrong call:
- False positive (flagged a page that's fine) → editor wastes ~15 minutes reviewing it. Low cost.
- False negative (missed a page that needs work) → opportunity lost. Harder to measure but compounds over time.

Because false positives are cheap and false negatives are invisible, the safe design is to cast a wider net and let the human decide — precision@K for a manageable queue size.
